```python
#task1
from functools import reduce
orders = [
    {"customer": "Alice", "total": 250.5},
    {"customer": "Bob", "total": "invalid_data"},
    {"customer": "Charlie", "total": 450},
    {"customer": "Daisy", "total": 100.0},
    {"customer": "Eve", "total": -30},  # Invalid total
]


def validate_orders(orders):
    # Function to check if total is valid ie.,numeric and>=0
    def is_valid(order):
        try:
            #Try to convert total to a float and check if it is >=0
            return float(order["total"]) >= 0
        except (ValueError, TypeError):
            # If fails, it is invalid
            return False

    # Using filter() and lambda to apply validation
    valid_orders = list(filter(lambda order: is_valid(order), orders))

    return valid_orders


valid_orders = validate_orders(orders)
print("Validate orders:")
print(valid_orders)


def apply_discount(orders):
    # Lambda function inside map to apply a 10% discount if total >300
    #The lambda checks if the total is greater than 300. If it is, it applies a 
    #10% discount (multiplies the total by 0.9).
    #{**order, "total": order["total"] * 0.9} creates a new dictionary based 
    #on the original order, but it has new total for theorders.

    discounted_orders = list(map(lambda order:
                                 {**order, "total": order["total"] * 0.9} if order["total"] > 300 else order, orders))

    return discounted_orders


discounted_orders = apply_discount(valid_orders)
print()
print("Discounted orders:")
print(discounted_orders)


def calculate_total_sales(orders):
    # Using reduce to sum up the 'total' of each order
    #The lambda function takes two arguments acc (accumulator) and 
    #order, where acc holds the running total, and order['total'] is the value added to acc.
    #the third argument 0 means the acc is initialized as 0
    total_sales = reduce(lambda acc, order: acc + order['total'], orders, 0)

    return total_sales


total_sales = calculate_total_sales(discounted_orders)
print()
print("Total sales:")
total_sales
```

class SquareIterator:
    def __init__(self, n):
        self.n = n
        self.current = 1

    def __iter__(self): #makes the class iterable by returning itself
        return self

    def __next__(self): # responsible for the iteration
        if self.current <= self.n:
            result = self.current ** 2
            self.current += 1
            return result
        else:
            raise StopIteration

square_iter = SquareIterator(6)
print("Squareiterator:")
for square in square_iter:
    print(square)


def fibonacci_generator(n):
    #taking a list to store the fibanonacci series
    fib_sequence = []
    a, b = 0, 1
    while a <= n:
        fib_sequence.append(a)
        a, b = b, a + b  #In loop a becomes b and for the value of b a is added in b 
    return fib_sequence

print()
print("fibonacci series:")
for num in fibonacci_generator(10):
    print(num)


```python
#task3
class DivisionError(Exception):
    """Custom exception for division-related errors."""
    pass


def divide_numbers(numbers, divisor):
    try:
        if divisor == 0:
            raise DivisionError("Divisor cannot be zero.")  # Raise custom exception

        results = []  
        for number in numbers:
            try:
                results.append(number / divisor) 
            except TypeError as e:
                # Raise a new DivisionError for non-numeric input with context
                raise DivisionError(f"Non-numeric input: {number}") from e

        return results 

    except DivisionError as e:
        print(f"Error: {e}")


# Example usage
numbers1 = [10, 20, 'a', 40] # List with a non-numeric input
numbers2 = [30, 40, 50] # list with correct inputs
print(divide_numbers(numbers1, 0))  # Divisor is zero
print(divide_numbers(numbers1, 2))  #error free divison
print(divide_numbers(numbers2, 2))  # Includes non-numeric input

print()
print()

def exception_logger(func):
    """Decorator to log exceptions raised in the wrapped function."""
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)  #  it calls the original function
        except Exception as e:

            print(f"Exception occurred in {func.__name__}: {type(e).__name__} - {e}")
    return wrapper

# Example usage
@exception_logger
def risky_divide(x):
    return 10 / x

print(risky_divide(2))
print(risky_divide(0))
print(risky_divide('a'))
```
